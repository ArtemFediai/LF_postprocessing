plot mobilities and current density


```python
def lineplot(data,
             save_to='mobility.png',
             y='mobility [cm^2/(V*sec)]'):
    sns_plot = sns.lineplot(data=data,
                            x='doping',
                            y=y,
                            markers='o',
                            style='field')
    sns_plot.set(xscale='log')
    sns_plot.set(yscale='log')
    plt.savefig(save_to, dpi=600)
    plt.close()  #<-- otherwise, it has a memory...
```

```python
# 1
lineplot(data=df_mobility)

# 2
lineplot(data=df_mobility,
         save_to='current_density.png',
         y='current density [A/m^2]')
```
